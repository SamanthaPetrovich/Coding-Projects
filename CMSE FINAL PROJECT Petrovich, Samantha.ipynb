## What is the most often used method to find planets

In this part of the data analysis I will be exploring the methods that most astronomers use to find planets. A few different types of mothods that are used are the transit method, the radial velocity method, the direct imaging method, the eclipse method, and the microlensing method. Let's explain the types of methods really quick:
* With the transit method astronomers look at the brightness of a particular star and they normalize it so it equals 1. When a planet moves in front of the star the brightness dips and depending on how deep the dip is and how wide the dip is will tell you how big the planet is and how log it orbits for. 
* With the radial velocity method astronomers look for a wiggle in the star's light. This wiggle shows that the star is being pulled by something and it shifts the light spectrum of the star to either blue-shifted or red-shifted. Depending on how large the shift is will tell you how large the planet is. 
* With the direct imaging method, it's pretty much you take a picture and look to see if there is a planet in front of the star. Direct imaging can only be done to stars that are nearby and not hudreds of light years away. 
* The eclipse method is used when there is a binary star system and looking for the offset of the eclipse from the planet that is pulling on it's host star.
* The last method that I want to talk about is the microlensing method. Microlensing occurs when there is two stars passing each other and the gravity from the star closest to us creates a lense that magnifies the star light behind it. The gravitational lense is able to bend the light from behind the star and intensify the effect of the light coming from the star. Since the light is intensified we are able to look at the dips in the light to see if there are any planets orbiting the star. This method is relatively new and is still developing.

## Making an example of how astronomers analyze planetary data from the trasit method

Since the transit method is the most popular way for Astronomers to find planets, I am making a simple example of how astronomers break down the data that they got and finding the orbit of the planet and how large the planet is.

```python
in_array = np.genfromtxt('keplerD_lc.dat')
times = in_array[:,0]
flux = in_array[:,1]
time = times - times[0] # here I am normalizing the time so that it starts from 0 and continues on. 
plt.plot(time,flux,',')
plt.xlabel('time in days')
plt.ylabel('Relative brightness')
```

This first part is loading in the data and taking the flux data and normalizing the time and then plotting the data. As you can see  there are dips that happen in the data and those are the transit dips from the planet. 

```python
plt.plot(time,flux)
timebetween = np.arange(3.84,45,4.88)
plt.vlines(3.84,1,0.993)
plt.vlines(8.72,1,0.993)
plt.vlines(timebetween,1,0.993)
plt.xlabel('time in days')
plt.ylabel('Relative brightness')
```

This part of the code is where you figure out the length of the transit and to see how long it takes for the planet to orbit the host star. Finding the amount of time between the dips in the brightness helps to determine how fast the planet is orbiting around the star.

```python
period = 4.8857
t_fold = np.remainder(time,period)#this is folding all of the dips in the data into one single dip. 
plt.scatter(t_fold,flux)#this is plotting the fold to the flux of the star
plt.xlim(3.8,4.15)
plt.ylim(0.991,1.005)
plt.xlabel('time in days')
plt.ylabel('Relative brightness')
```

With this code we are folding the data onto itself to show the averaged dip in the relative brightness of the star. This is an important step especially getting the period of the transit right, so that you can get a good look at the brightness.

```python
period = 4.8857
t_fold = np.remainder(time,period)
plt.scatter(t_fold,flux)
plt.xlim(3.8,4.15)
plt.ylim(0.991,1.002)
plt.hlines(0.99255,4.09,3.85)
plt.hlines(1,4.09,3.85)
depth = (1-0.99255)*100 
regdepth = (1-0.99255)
plt.xlabel('time in days')
plt.ylabel('Relative brightness')
print('percentage of depth',depth)
print(regdepth)
```

Here I am finding the depth of the dip in brightness so that I can better determine the size of the planet from the percentage of the light that is blocked off. In this part I am putting the horizontal lines int he best fit spot on the top and the bottom of the curve.

```python
period = 4.8857 #in days
t_fold = np.remainder(time,period) #this is folding all of the dips in the data into one single dip. 
plt.scatter(t_fold,flux) #Plotting the fold of the data to the flux
plt.xlim(3.8,4.15)
plt.ylim(0.991,1.002)
plt.vlines(3.85,1.001,0.992)
plt.vlines(4.078,1.001,0.992)
plt.xlabel('time in days')
plt.ylabel('Relative brightness')
trandur = 4.08-3.85
print('The duration of the planet in front of the star is',trandur,'days')
```

Here I am finding the duration of the transit in order to see how long it is passing in front of the star.

```python
hoursofdur = trandur*24
print('The amount of time, in hours, that the transit was in front of the star was',hoursofdur,'.')
```

This bit of code is just converting the duration of the transit from days to hours so we have a better understanding of how long the transit is in front of the star.
